```python
#c) Observa cuántos Nas tienes por columna y por fila(usa la función que vimos para hacerlo para todas las columnas/filas
#de una vez). Elimina las filas en donde haya valores nulos.

def num_nulos(x):
  return x.isnull().sum()

print("Número de Nas por fila en el datset train: \n")
print("Nas por fila:")
print(train.apply(num_nulos, axis=1))

print("\n")
print("Número de Nas por columna en el dataset train: \n")
print(train.apply(num_nulos, axis=0))

print("Número de Nas por fila en el datset test:\n ")
print("Nas por fila:")
print(test.apply(num_nulos, axis=1))

print("\n")
print("Número de Nas por columna en el dataset test: ")
print(test.apply(num_nulos, axis=0))

print("Sustituyo Nas con otro valor y verificacion de cero nas en train: ")
train.fillna(0, inplace=True)
print(train.apply(num_nulos, axis=1))

print("Sustituyo Nas con otro valor y verificacion de cero nas en test: ")
test.fillna(0, inplace=True)
print(test.apply(num_nulos, axis=1))
```

```python
#d) Recodifica a 0-1 las variables categóricas que estén siendo tratadas como strings. 
#Usa la función replace que vimos, con un diccionario para hacer esto.

test2=test.replace({'Married': "Yes", 'Self_Employed': "Yes", 'Education': "Graduate", 'Loan_Status':"Y"}, 1)
test3=test2.replace({'Married': "No", 'Self_Employed': "No", 'Education': "Not Graduate", 'Loan_Status':"N"}, 0)
test3["Married"] = train3["Married"].astype("category")
test3["Self_Employed"] = train3["Self_Employed"].astype("category")
test3["Education"] = train3["Education"].astype("category")
test3["Loan_Status"] = train3["Loan_Status"].astype("category")
print(test3)

train2=train.replace({'Married': "Yes", 'Self_Employed': "Yes", 'Education': "Graduate"}, 1)
train3=train2.replace({'Married': "No", 'Self_Employed': "No", 'Education': "Not Graduate"}, 0)
train3["Married"] = train3["Married"].astype("category")
train3["Self_Employed"] = train3["Self_Employed"].astype("category")
train3["Education"] = train3["Education"].astype("category")
train3["Loan_Status"] = train3["Loan_Status"].astype("category")
print(train3)
```
