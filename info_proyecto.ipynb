#### MASTER DATA MANAGEMENT Y DATOS MAESTROS

MDM es el conjunto de metodologías, herramientas y procesos, necesarios para crear y mantener conjuntos precisos y consistentes de datos maestros. De esta forma se identifica la información más importante de una empresa, creando una única fuente de la verdad, que permite a la organización mejorar sus procesos empresariales.

Dato Maestro
Datos críticos de un negocio que permiten identificar personas, objetos, lugares y conceptos. MDM permite gestionar el conjunto de estos datos maestros de forma que no haya inconsistencias entre ellos.

##### 2. Reglas

- Número de Cliente
Sin combinar con otro dato: Confidencial
Combinado con otro datos: Confidencial PII (personal identificable Information).
Número de identificación del registro del cliente con el banco. 
Es proporcionado de manera automática durante el proceso de Alta del cliente o del tercero relacionado.
Es el dato que permite relacionar al cliente en todos sus productos y servicios.
Identifica de manera única el registro del cliente con productos y servicios, y es llave única de todas las estructuras del sistema de clientes.
Permite identificar terceros relacionados, aún y cuando no mantengan una relación directa de negocio con el banco.
Permite asociar al cliente con sus datos demográficos (nombre, domicilio, teléfonos, etc.)

- Nombre
Dato Confidencial PII
Se refiere al nombre completo (nombre(s), apellido paterno y materno cuando aplique) de la persona física o la razón social para persona moral. Es el principal dato de identificación e identidad, proporcionado por el cliente.
Se permite el uso de los siguientes caracteres y símbolos:
    •	PF/PFAE:
Letras (A-N, O-Z), espacio y arroba (@) para representar a la letra Ñ. Como separadores de nombre(s), apellido paterno y apellido materno se admiten los caracteres especiales coma (,) y diagonal (/).
No se permiten abreviaturas.
    •	PM:
Letras (A-N, O-Z), espacio, arroba (@) para representar a la letra Ñ, dígitos (0123456789), diagonal (/), guion medio (-) y ampersand (&).
Para personas morales el nombre debe corresponder al acta constitutiva.

- Domicilio

El dato Domicilio es proporcionado por el cliente y se refiere al lugar físico en donde se le puede localizar o contactar. 
El domicilio está conformado por:
•  Calle y número.
•  Colonia.
•  Población (Municipio para domicilios del interior de la República Mexicana o Delegación Política para la Ciudad de México).
•  Código Postal.
•  Entidad Federativa o Estado.
El Domicilio Principal es aquél que el cliente define como su lugar de residencia. Adicionalmente es el que emplea para fines fiscales ante las autoridades correspondientes.

- Población
Dato Confidencial PII
Nombre propio que identifica al Municipio y en el caso de la Ciudad de México a la Delegación Política del domicilio principal o de correspondencia del cliente.

- Correo electrónico
Es el nombre asociado a un buzón de correo electrónico a través de la cual es factible establecer comunicación electrónica escrita con el Cliente usando texto, imágenes y archivos a través de los medios electrónicos disponibles actualmente.
El estándar que norma el nombre del correo electrónico es definido por el IETF a nivel internacional, para garantizar la interoperabilidad de plataformas tecnológicas.
Reglas:
•	No nulo, no espacio, longitud mayor a 1.
•	Contiene "@" y un dominio
•	Longitud mayor a 4. (Ej. a@b.c)
•	No se permiten espacios intermedios
•	No es válido ningún caracter especial al inicio. Ej: .jfieo@hotmail.com, =rosll@hotmail.com
•	No es válido ningún caracter especial inmediatamente antes o después de la arroba(@). Ej jdlioii/@_hotmail.com, kdoelkj.@gmailc.om.
•	Longitud máxima 254 para efectos de forward/reverse. La tolerancia en longitud para local-part y domain-part es {64}@{255} respectivamente. IETF (Internet Engineering Task Force)
•	Una dirección entregable de email es una dirección de corree electrónico que puede ser usada para entregar mensajes sobre una infraestructura SMTP(Simple Mail Transfer Protocol) moderna. El patrón localName@internetDomain lo norma el IETF (Internet Engineering Task Force) para garantizar la interoperabilidad al enviar/recibir correos al/del cliente sin importar el servidor o servicio de correo electrónico que se tenga. Los estándares más relevantes son RFC5321 y RFC5322. El draft IETF en https://tools.ietf.org/html/draftseantek- mail-regexen-00 válido al September 22, 2016 propone validar requisitos de estándares de SMTP (RFC5321 y RFC6531), de principios DNS (RFC1034, RFC1035, RFC1123) y de IDNA moderno (RFC5890, RFC5891, RFC5892) con el siguiente RegEx (regular expression): P E J DEA = ([A-Za-z0-9!#-'*+\-/=?^_`{- ~\xA0-\x{10FFFF}]+ (?:\.[A-Za-z0-9!#-'*+\-/=?^_`{-~\xA0-\x{10FFFF}])|"(?:[ !#-\[\]-\xA0-\x{10FFFF}]|\\[ -~])*")@((?:[A-Za-z0-9](?:[A-Za-z0-9\-]*[A-Za-z0-9])?|[\x00-\x{10FFFF}]*[\x80-\x{10FFFF}]+[\x00-\x{10FFFF}]*)(?:\.(?:[A-Za-z0-9](?:[A-Zaz0-9\-]*[A-Za-z0-9])?|[\x00-\x{10FFFF}]*[\x80-\x{10FFFF}]+[\x00- \x{10FFFF}]*))*)Probada en PCRE (Perl Compatible RegularExpressions) 8.38, en PERE (POSIX Extended Regular Expressions) POSIX/IEEE Std 1003.1, enJSRE (JavaScript Regular Expressions)ECMAScript/ECMA-262 y en std::regex library de C++11.
•	Email no asignado a más de 4 clientes diferentes.

- Teléfono

Es un dato de contacto con el Cliente, la comunicación se establece de forma oral o escrita y directa a través de la tecnología que actualmente existe para tal fin, puede ser fijo o móvil.
Un cliente puede tener registrado más de un número de teléfono por lo que es necesario tener clasificado como teléfono de casa, personal, de oficina o de referencia.
Reglas:

•	No nulo, no vacío, no cero, no menor a cero.
•	Patrón <1 digito diferente de 0>+<9 dígitos>. Si el campo es tipo numérico en la fuente entonces puede validarse que sea mayor a 1,000,000,000 para cumplir la regla. Ayuda también a que excluyan prefijos de acceso como son 01, 044, 045, +1.
•	No debe ser teléfono que corresponda a ejecutivos u operadores del Banco o sus filiales. Se permite solo si el teléfono se proporciona en  número de cliente del empleado y como teléfono de oficina.
•	A partir del número telefónico debe consultarse el catálogo IFT(antes COFETEL) para conocer si es FIJO o MOVIL, y debe corresponder con el tipo declarado.
•	No se permite un digito repetido siempre en todo el número. Ejemplo 10 ceros, 10 unos, etc.
